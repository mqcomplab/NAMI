# Text File to CSV Converter

This notebook converts text files to CSV format. It supports various delimiters (space, tab, comma, custom).

**Instructions:**
1. Update the `input_file` variable with your text file path
2. Adjust the delimiter if needed
3. Run all cells to generate the CSV file

## 1. Import Required Libraries

````python
import pandas as pd
import os
````

## 2. Configure File Paths and Delimiter

````python
# Specify input text file path
input_file = "smiles_all_20.txt"  # Change this to your text file path

# Specify delimiter (choose one or specify custom)
# Common options: ' ' (space), '\t' (tab), ',' (comma), '|' (pipe)
delimiter = ' '  # Default: space-separated

# Specify if file has header row
has_header = False  # Set to True if first row contains column names

# Output CSV file path (optional - defaults to same name with .csv extension)
output_file = None  # Will auto-generate if None
````

## 3. Read and Parse the Text File

````python
# Check if file exists
if not os.path.exists(input_file):
    print(f"Error: File '{input_file}' not found!")
    print(f"Current directory: {os.getcwd()}")
else:
    # Read the text file
    if has_header:
        df = pd.read_csv(input_file, sep=delimiter, engine='python')
    else:
        df = pd.read_csv(input_file, sep=delimiter, header=None, engine='python')
    
    print(f"Successfully loaded {len(df)} rows and {len(df.columns)} columns")
    print(f"\nData types:\n{df.dtypes}")

# Column 1 name is smiles and column 2 name is dataset
df.columns = ['smiles', 'dataset']
````

## 4. Preview the Data

````python
# Display first few rows
print("First 10 rows:")
print(df.head(10))

# Display basic statistics
print("\n" + "="*50)
print("Data Summary:")
print("="*50)
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"\nColumn names: {list(df.columns)}")
````

## 5. (Optional) Add Column Names

If your text file doesn't have headers, you can add custom column names here.

````python
# Optional: Rename columns if needed
# Uncomment and modify the line below:
# df.columns = ['Column1', 'Column2', 'Column3']  # Replace with your column names

# Display updated column names
print(f"Current columns: {list(df.columns)}")
````

## 6. Export to CSV

````python
# Generate output filename if not specified
if output_file is None:
    base_name = os.path.splitext(input_file)[0]
    output_file = f"{base_name}.csv"

# Export to CSV
df.to_csv(output_file, index=False)

print(f"✓ CSV file successfully created: {output_file}")
print(f"✓ Saved {len(df)} rows and {len(df.columns)} columns")
````

## Examples of Different Text File Formats:

**Space-separated:**
```python
input_file = "data.txt"
delimiter = ' '
```

**Tab-separated:**
```python
input_file = "data.txt"  
delimiter = '\t'
```

**Comma-separated:**
```python
input_file = "data.txt"
delimiter = ','
```

**Custom delimiter (e.g., pipe):**
```python
input_file = "data.txt"
delimiter = '|'
```

**Multiple spaces as delimiter:**
```python
input_file = "data.txt"
delimiter = '\s+'  # Regex pattern for one or more spaces
```
